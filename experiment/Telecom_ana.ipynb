```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def uniplot(df, col, title, hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    fig, ax = plt.subplots()
    unique_col_count = len(df[col].unique())
    unique_hue_count = len(df[hue].unique()) if hue else 0
    width = unique_col_count + 7 + 4 * unique_hue_count
    fig.set_size_inches(width, 8)
    
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)

    # Conditional hue parameter
    if hue is not None:
        ax = sns.countplot(data=df, x=col, order=df[col].value_counts().index, hue=hue, palette='bright')
    else:
        ax = sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='bright')
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```
